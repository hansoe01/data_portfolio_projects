Question:4(b) Predictive Model for Tips Percentage of the Total Fares

```python
# Data Preparation for ML Model

# Ensure no division issues
df = df[df['total_amount'] > 0]

# Target variable
df['tip_pct'] = df['tip_amount'] / df['total_amount']

# Remove extreme / noisy values
df = df[(df['tip_pct'] >= 0) & (df['tip_pct'] <= 1)]
df = df[df['trip_distance'] > 0]


```

```python
#Feacture Engineering
# Time features
df['pickup_hour'] = df['lpep_pickup_datetime'].dt.hour
df['pickup_day'] = df['lpep_pickup_datetime'].dt.dayofweek

# Ensure lpep_dropoff_datetime is also a datetime object
df['lpep_dropoff_datetime'] = pd.to_datetime(df['lpep_dropoff_datetime'])

# Trip duration (minutes)
df['trip_time_min'] = (
    df['lpep_dropoff_datetime'] - df['lpep_pickup_datetime']
).dt.total_seconds() / 60

# Remove invalid durations
df = df[df['trip_time_min'] > 0]
```

```python
# Convert 'passenger_count' to numeric, coercing invalid entries to NaN
df['passenger_count'] = pd.to_numeric(df['passenger_count'], errors='coerce')

# Select relevant numeric features
corr_df = df[[
    'trip_distance',
    'fare_amount',
    'total_amount',
    'passenger_count',
    'trip_time_min',
    'pickup_hour',
    'pickup_day',
    'tip_pct'
]].copy()

# Remove missing values
corr_df = corr_df.dropna()
```

```python
from sklearn.model_selection import train_test_split

# Define X (features) and y (target)
X = corr_df.drop('tip_pct', axis=1)
y = corr_df['tip_pct']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,   # 20% test
    train_size=0.8,  # 80% train
    random_state=42
)

print("Training Set Size:", X_train.shape)
print("Test Set Size:", X_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=100,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("\nModel Evaluation (Test Set):")
print("MAE:", mae)
print("RMSE:", rmse)
print("R² Score:", r2)
```

```python
results = pd.DataFrame({
    'Actual': y_test.values[:10],
    'Predicted': y_pred[:10]
})

print("\nSample Predictions:")
print(results)
```

```python

plt.figure()

plt.scatter(y_test, y_pred, alpha=0.5)

plt.xlabel("Actual Tip %")
plt.ylabel("Predicted Tip %")
plt.title("Actual vs Predicted Tip Percentage")

# Perfect prediction line
plt.plot([0,1], [0,1])

plt.show()
```

## Question:5_ Distributions Analysis related to Average Speed

Question:5(a) Build a derived variable representing the average speed over the course of a trip.

```python
# Trip duration in hours
df['trip_duration_hr'] = (df['lpep_dropoff_datetime'] - df['lpep_pickup_datetime']).dt.total_seconds() / 3600

# Average speed (mph)
df['avg_speed'] = df['trip_distance'] / df['trip_duration_hr']

# Clean unrealistic values
df = df[(df['avg_speed'] > 0) & (df['avg_speed'] < 80)]
```

```python
df['week'] = df['lpep_pickup_datetime'].dt.isocalendar().week
df['month'] = df['lpep_pickup_datetime'].dt.month

# Filter September
sept_df = df[df['month'] == 9]
```

```python
sns.boxplot(x='week', y='avg_speed', data=sept_df)
plt.title("Average Speed by Week (September)")
plt.show()
```

```python
hourly_speed = df.groupby('pickup_hour')['avg_speed'].mean()

print("Average Speed by Hour:")
print(hourly_speed)

hourly_speed.plot(title="Average Speed by Hour")
plt.show()
```

Hypothesis

“Average trip speed follows a U-shaped pattern across the day, with lowest speeds during peak commuting hours due to congestion, and highest speeds during early morning and late-night hours when traffic volume is minimal.”

Question:5(b) Can you perform a test to determine if the average trip speeds are materially the same in all weeks of September? If you decide they are not the same, can you form a hypothesis regarding why they differ? Can you build up a hypothesis of average trip speed as a function of time of day?

```python
# To test the average speeds are the same in all weeks of September
from scipy.stats import f_oneway

groups = [group['avg_speed'].values for name, group in sept_df.groupby('week')]

f_stat, p_value = f_oneway(*groups)

print("F-stat:", f_stat)
print("p-value:", p_value)
```

```python
df['hour'] = df['lpep_pickup_datetime'].dt.hour
hourly_speed = df.groupby('hour')['avg_speed'].mean().reset_index()
```

```python
plt.figure(figsize=(10,6))
plt.plot(hourly_speed.index, hourly_speed.values, marker='o')
plt.xlabel("Hour of Day")
plt.ylabel("Average Speed (mph)")
plt.title("Average Trip Speed by Hour of Day")
plt.grid(True)
plt.show()
```

Hypothesis: Weekly variation in average speed is likely driven by external factors such as weather conditions, demand surges, and traffic congestion variability.
